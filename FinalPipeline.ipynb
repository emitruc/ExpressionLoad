```python
def select_gene(scaf, x):
    efs = ['synonymous', 'missense']
    if scaf.empty == True:
        res = [0,0,0,0]
    else:
        res = []
        for i in efs:
            res.append(len(scaf[scaf['vartype']==i]['derived'+x]))
            res.append(scaf[scaf['vartype']==i]['derived'+x].sum())
    return res
```

```python
data = pd.read_table('daf.joint.no00', header=[0])
genes = pd.read_table('geneCoord', header=[0])
```

```python
genesList = genes['gene'].to_list()

gd = pd.DataFrame(columns=['gene','eSyn#', 'eSyns','eMis#', 'eMiss',
                           'kSyn#', 'kSyns', 'kMis#', 'kMiss'])


data1 = data[(data['avgCov']>6)&
             (data['avgCov']<8)&
             (data['total_out']>3)&
             (data['flagPol']=='unfolded')]

###
###select only private segregating mutations

seg_e = data1[(data1['total1']==48)&
             (data1['derived1']>0)&
             (data1['derived1']<48)&
             (data1['derived2']==0)]

seg_k = data1[(data1['total2']==48)&
              (data1['derived2']>0)&
              (data1['derived2']<48)&
              (data1['derived1']==0)]
###


###select only fixed differences (but calling them seg...)
#
#seg_e = data1[(data1['total1']==48)&
#             (data1['derived1']==48)&
#             (data1['derived2']==0)]

#seg_k = data1[(data1['total2']==48)&
#              (data1['derived2']==48)&
#              (data1['derived1']==0)]
###

l = len(genesList)
i = 0
for gene in genesList:
    print(round((i/l)*100,2))
    scaf = genes[genes['gene']==gene]['scaffold'].values[0]
    start = genes[genes['gene']==gene]['start'].values[0]
    end = genes[genes['gene']==gene]['end'].values[0]
    
    e = seg_e[(seg_e['scaffold']==scaf)&
                (seg_e['position']>start-1)&
                (seg_e['position']<end+1)]
    
    k = seg_k[(seg_k['scaffold']==scaf)&
                (seg_k['position']>start-1)&
                (seg_k['position']<end+1)]

    gene_data = [gene]

    gene_data += select_gene(e, '1')
    gene_data += select_gene(k, '2')
   
    gd.loc[0 if pd.isnull(gd.index.max()) else gd.index.max() + 1] = gene_data
    i += 1

gd.to_csv('genesSegCounts_6-8', header=True, index=None, sep='\t')
```

```python
#TPM normalization by CDS length (kb); it can also be done by exon length if you want
df['eRPK'] = df['eExp']/(df['CDSlen']/1000)
df['kRPK'] = df['kExp']/(df['CDSlen']/1000)
eTPMscal = df.eRPK.sum()/1000000
kTPMscal = df.kRPK.sum()/1000000
df['eTPM'] = df['eRPK']/eTPMscal
df['kTPM'] = df['kRPK']/kTPMscal

#Polymorphic sites normalization by CDS lenghts (kb)

df['eSyn#_n']=df['eSyn#']/(df['CDSlen']/1000)
df['eSyns_n']=df['eSyns']/(df['CDSlen']/1000)
df['eMis#_n']=df['eMis#']/(df['CDSlen']/1000)
df['eMiss_n']=df['eMiss']/(df['CDSlen']/1000)

df['kSyn#_n']=df['kSyn#']/(df['CDSlen']/1000)
df['kSyns_n']=df['kSyns']/(df['CDSlen']/1000)
df['kMis#_n']=df['kMis#']/(df['CDSlen']/1000)
df['kMiss_n']=df['kMiss']/(df['CDSlen']/1000)

df['epN/pS'] = df['eMis#']/df['eSyn#']
df['kpN/pS'] = df['kMis#']/df['kSyn#']

df['epN/pS_n'] = df['eMis#_n']/df['eSyn#_n']
df['kpN/pS_n'] = df['kMis#_n']/df['kSyn#_n']

df.replace(np.inf, np.nan, inplace=True)
```

```python
### Import SFS after removing the intergenic SNPs from the daf.joint.no00.rnaCov with a simple bash command: 
### awk '$14 != "intergenic"' daf.joint.no00.rnaCov > daf.joint.no00.rnaCov.noIntergenic
### then normalize expression and apply basic filters

data = pd.read_table('daf.joint.no00.rnaCov.noIntergenic', header=[0])
data[['rnaCov1','rnaCov2']] = data[['rnaCov1','rnaCov2']].fillna(0)

#CPM normalization
eCPMscal = data.rnaCov2.sum()/1000000
kCPMscal = data.rnaCov1.sum()/1000000
data['eCPM'] = data['rnaCov2']/eCPMscal
data['kCPM'] = data['rnaCov1']/kCPMscal

#CPM Z-normalisation just for easiness of interpretation (it does not change the distribution!)
data['eCPMstdz'] = (data['eCPM'] - data['eCPM'].mean()) / data['eCPM'].std()
data['kCPMstdz'] = (data['kCPM'] - data['kCPM'].mean()) / data['kCPM'].std()


#basic filters on coverage and missingness as before
data1 = data[(data['avgCov']>6)&
             (data['avgCov']<8)&
             (data['total_out']>3)&
             (data['flagPol']=='unfolded')]
```

```python
###Select emp or king private fixed and segregating derived mutations and write to separate files
segderE = data1[(data1['total1']==48)&
             (data1['derived1']>0)&
             (data1['derived2']==0)]

segderK = data1[(data1['total2']==48)&
              (data1['derived2']>0)&
              (data1['derived1']==0)]

LOW_E = segderE[(segderE['effect']=='LOW')&
                (segderE['vartype']=='synonymous')]

LOW_K = segderK[(segderK['effect']=='LOW')&
                (segderK['vartype']=='synonymous')]

MDR_E = segderE[(segderE['effect']=='MODERATE')]

MDR_K = segderK[(segderK['effect']=='MODERATE')]

HIGH_E = segderE[(segderE['effect']=='HIGH')]

HIGH_K = segderK[(segderK['effect']=='HIGH')]

LOW_E.to_csv('empLOWsynlist.vf', header=True, index=None, sep='\t')
LOW_K.to_csv('kingLOWsynlist.vf', header=True, index=None, sep='\t')

MDR_E.to_csv('empMDRlist.vf', header=True, index=None, sep='\t')
MDR_K.to_csv('kingMDRlist.vf', header=True, index=None, sep='\t')

HIGH_E.to_csv('empHIGHlist.vf', header=True, index=None, sep='\t')
HIGH_K.to_csv('kingHIGHlist.vf', header=True, index=None, sep='\t')
```
